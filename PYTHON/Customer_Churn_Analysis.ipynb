# Exploratory Data Analysis (EDA)

In this section, we analyze customer demographics, service usage, contract types, and churn patterns to identify the major factors influencing customer churn.

```python
churn_counts = df["Churn"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(churn_counts.index, churn_counts.values)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

Most customers did not churn, indicating that customer retention is generally strong. However, a significant number of customers have left, making churn prediction an important business problem.

```python
contract = df.groupby("Contract")["customerID"].count()

plt.figure(figsize=(7,5))
plt.bar(contract.index, contract.values)

plt.title("Customers by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Customers")

plt.show()
```

### Insight

Month-to-Month contracts represent the largest customer segment. These customers generally have a higher likelihood of churn compared to customers with long-term contracts.

```python
internet = df.groupby("InternetService")["customerID"].count()

plt.figure(figsize=(7,5))
plt.bar(internet.index, internet.values)

plt.title("Customers by Internet Service")

plt.xlabel("Internet Service")
plt.ylabel("Customers")

plt.show()
```

### Insight

Fiber Optic and DSL services account for most customers. Further analysis will determine which service experiences higher churn.

```python
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)

plt.title("Customer Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")

plt.show()
```

### Business Insight

Customers with **Month-to-Month** contracts have the highest churn. Customers on **One Year** and **Two Year** contracts are much more likely to stay, suggesting that longer commitments improve customer retention.

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x="InternetService",
    hue="Churn"
)

plt.title("Customer Churn by Internet Service")
plt.xlabel("Internet Service")
plt.ylabel("Number of Customers")

plt.show()
```

### Business Insight

Customers using **Fiber Optic** services experience noticeably higher churn than DSL users. This may indicate pricing, service quality, or customer expectation issues.

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="PaymentMethod",
    hue="Churn"
)

plt.xticks(rotation=20)

plt.title("Customer Churn by Payment Method")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")

plt.show()
```

### Business Insight

Customers paying via **Electronic Check** show the highest churn, while customers using automatic payment methods tend to remain with the company longer.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Churn",
    y="MonthlyCharges"
)

plt.title("Monthly Charges vs Customer Churn")

plt.show()
```

### Business Insight

Customers who churn generally have higher monthly charges compared to retained customers. This suggests that pricing could be one of the major reasons behind customer churn.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="Churn",
    y="tenure"
)

plt.title("Customer Tenure vs Churn")

plt.show()
```

### Business Insight

Customers with lower tenure are more likely to churn. Long-term customers tend to remain loyal to the company.

```python
plt.figure(figsize=(7,5))

sns.countplot(
    data=df,
    x="SeniorCitizen",
    hue="Churn"
)

plt.title("Senior Citizens and Customer Churn")

plt.show()
```

### Business Insight

Senior citizens show a different churn pattern compared to other customers, indicating that age-specific retention strategies may be beneficial.

```python
df_corr = df.copy()
```

```python
binary_cols = [
    "Partner","Dependents","PhoneService",
    "PaperlessBilling","Churn"
]

for col in binary_cols:
    df_corr[col] = df_corr[col].map({"Yes":1,"No":0})
```

```python
numeric_df = df_corr.select_dtypes(include=["int64","float64"])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")

plt.show()
```

# Final Business Recommendations

## Key Findings

- Month-to-Month customers have the highest churn.
- Fiber Optic customers churn more frequently.
- Electronic Check users have the highest churn.
- Customers with higher monthly charges are more likely to leave.
- Customers with shorter tenure have the highest churn.

## Business Recommendations

- Promote long-term contracts through discounts.
- Review Fiber Optic pricing and service quality.
- Encourage customers to switch to automatic payment methods.
- Focus retention campaigns on new customers during their first year.
- Offer personalized retention plans for high-value customers.
